```python
import pandas as pd
import matplotlib.pyplot as plt

# Creazione del dataset simulato
data = {
    "Nome": ["Hotel Inclusivo Roma", "B&B Accessibile Firenze", "Agriturismo Senza Barriere Napoli",
             "Hotel Accessibile Milano", "B&B Accogliente Torino"],
    "Posizione": ["Lazio, Roma", "Toscana, Firenze", "Campania, Napoli", "Lombardia, Milano", "Piemonte, Torino"],
    "Servizi": ["Rampe, Ascensore, Bagni accessibili", "Ascensore, Guide tattili",
                "Rampe, Servizi per ipovedenti", "Ascensore, Bagni accessibili", "Rampe, Guide tattili"],
    "Tipo": ["Hotel", "B&B", "Agriturismo", "Hotel", "B&B"],
    "Valutazione": [4.5, 4.7, 4.3, 4.6, 4.4]
}

# Caricamento del DataFrame
df_cleaned = pd.DataFrame(data)

# Analisi: Valutazione Media per Tipo di Struttura
valutazione_media_tipo = df_cleaned.groupby("Tipo")["Valutazione"].mean()

# Grafico: Valutazione Media per Tipo di Struttura
plt.figure(figsize=(8, 5))
valutazione_media_tipo.plot(kind="bar", color="skyblue", title="Valutazione Media per Tipo di Struttura")
plt.ylabel("Valutazione Media")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analisi: Distribuzione delle Strutture per Regione
df_cleaned['Regione'] = df_cleaned['Posizione'].apply(lambda x: x.split(',')[0])
distribuzione_regione = df_cleaned['Regione'].value_counts()

# Grafico: Distribuzione delle Strutture per Regione
plt.figure(figsize=(8, 5))
distribuzione_regione.plot(kind="bar", color="lightgreen", title="Distribuzione delle Strutture per Regione")
plt.ylabel("Numero di Strutture")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analisi: Servizi pi Comuni
servizi = df_cleaned['Servizi'].str.split(', ').explode()
servizi_comuni = servizi.value_counts()

# Grafico: Servizi pi Comuni
plt.figure(figsize=(8, 5))
servizi_comuni.plot(kind="bar", color="salmon", title="Servizi Pi Comuni nelle Strutture")
plt.ylabel("Numero di Strutture")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
